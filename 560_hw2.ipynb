# step1

create 1000 random numbers over the range 0-100, each number is a two-digit decimal number, we store it in a txt file called "dataset_x.txt".

```python
import random
import matplotlib.pyplot as plt

dataset_x = []
output = "dataset_x.txt"
for i in range(0, 1000):
    num_100 = random.randint(1, 10000)
    dataset_x.append(num_100 / 100)
print(dataset_x)

plt.hist(dataset_x, bins=40, normed=0, facecolor="blue", edgecolor="black", alpha=0.7)
plt.show()
```

# step2

according to the dataset which create in step1 (dataset_x), we use equation y = 3x+6, (x is the nyumber in dataset_x) we create a new dataset called dataset_y and we store it in a txt file called "dataset_y.txt".

```python
dataset_y = []
files = open("dataset_x.txt","r")
for file in dataset_x:
    dataset_y.append(float(file)*3+6)
print(dataset_y)

plt.hist(dataset_y, bins=40, normed=0, facecolor="blue", edgecolor="black", alpha=0.7)
plt.show()
```

# step3

we use dataset_x and dataset_y to create a new graph, the index of the graph is the data in dataset_x, and the value of index get from dataset_y. Then we use line to connect each point to obtain the visulization graph. Also, we store it as a .jpg file.

```python
plt.figure(figsize=(7,5))
plt.scatter(dataset_x,dataset_y, label='o', color='red', s=3)
plt.plot(dataset_x,dataset_y,color="blue",linewidth = 2)
plt.xlabel("index")
plt.ylabel("value of index")
plt.title("Visualization")
plt.savefig("visualization.png")
plt.show()
```
